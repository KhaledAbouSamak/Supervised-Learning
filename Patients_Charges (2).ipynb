# Conducting a formal significance test for one of the hypotheses and discuss the results 



## **1. Significance  Test for the average charges: by Gender**


```python
#Calculate the average charges by gender and test if it is statistically significant
data.groupby('sex')['charges'].mean()
```

H0: no diffrence in mean charges between male and female

H1: There is significant difference in mean charges between male and female

```python

import pandas as pd
from scipy import stats

F, p = stats.f_oneway(data['charges'][data['sex']==1], data['charges'][data['sex']==0])
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
# ''''''
```

## **2. Significance  Test for the average charges: by Gender**


```python
#Calculate the average charges by smoker and test if it is statistically significant
data.groupby('smoker')['charges'].mean()
```

H0: no diffrence in mean charges between smoker and non smoker

H1: There is significant diffrence in mean charges between smoker and non smoker

```python
F, p = stats.f_oneway(data['charges'][data['smoker']==1], data['charges'][data['smoker']==0])
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
# ''''''
```

## **3. Significance  Test for the average charges: by Region**


```python
#Calculate the average charges by region and test if it is statistically significant
data.groupby('region')['charges'].mean()
```

H0: no diffrence in mean charges between regions

H1: There is significant difference in mean charges between regions

```python
F, p = stats.f_oneway(data['charges'][data['region']==0], data['charges'][data['region']==1], data['charges'][data['region']==2], data['charges'][data['region']==3])
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
# ''''''
```

# 3. Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
#sex
le = LabelEncoder()
le.fit(data.sex.drop_duplicates()) 
data.sex = le.transform(data.sex)
# smoker or not
le.fit(data.smoker.drop_duplicates()) 
data.smoker = le.transform(data.smoker)
#region
le.fit(data.region.drop_duplicates()) 
data.region = le.transform(data.region)
```

Numeric features scaling

Scale the age and bmi feratures

```python
data['age'] = (data['age'] - data['age'].mean())/data['age'].std()
data['bmi'] = (data['bmi'] - data['bmi'].mean())/data['bmi'].std()

data.head()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['zscore_bmi', 'charges'],axis = 1)
y = data['charges']
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = .3,random_state = 0)
x_train.shape, x_test.shape
```

##### Ploynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 2)
X_pf = pf.fit_transform(X)
print( 'the shape of data after adding pf', X_pf.shape)
x_train_pf, x_test_pf, y_train_pf, y_test_pf = train_test_split(X_pf,y, test_size = .3,random_state = 0)
x_train_pf.shape, x_test_pf.shape
```

## 4. Prediction

#### Linear Regression as base model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()
lr.fit(x_train,y_train)


r2_score(y_test, lr.predict(x_test))
```

#### Linear Regression with ploynomial features

```python
lr_pf = LinearRegression()
lr_pf.fit(x_train_pf,y_train_pf)

r2_score(y_test_pf, lr_pf.predict(x_test_pf))
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha = 1)
rr.fit(x_train,y_train)

r2_score(y_test, rr.predict(x_test))
```

#### Ridge Regression with ploynomial features

```python
rr_pf = Ridge(alpha = 1)
rr_pf.fit(x_train_pf,y_train_pf)

r2_score(y_test_pf, rr_pf.predict(x_test_pf))
```

```python
from sklearn.linear_model import Lasso

lar = Lasso(alpha = 1)
lar.fit(x_train,y_train)

r2_score(y_test, lar.predict(x_test))
```

```python
lar_pf = Lasso(alpha = 1)
lar_pf.fit(x_train_pf,y_train_pf)

r2_score(y_test_pf, lar_pf.predict(x_test_pf))
```

## Key Findings

I have built diffrent regression models using vanila linear regression, Lasso, Ridge with features and with Ploynomial features. Since linear model was not overfitting data, the experiment has shown that adding regularization has not improved the performance.
